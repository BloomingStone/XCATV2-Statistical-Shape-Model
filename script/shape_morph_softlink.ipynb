```python
from pathlib import Path
import nibabel as nib
import numpy as np
from tqdm import tqdm


output_ssm_nii_root = Path('/media/data3/sj/Data/Phatom/output_ssm_nii')
output_ssm_nii__with_vessels_root = Path('/media/data3/sj/Data/Phatom/output_ssm_nii__with_vessels')
shape_morph_root = Path('/media/data3/sj/Data/ShapeMorph')

for case_dir in tqdm(output_ssm_nii_root.iterdir(), desc='total'):
    label_dir = case_dir / 'label'
    for label_nii in tqdm(label_dir.glob('*.nii.gz'), desc='case'):
        name = label_nii.name
        image_vessel_nii = output_ssm_nii__with_vessels_root / case_dir.name / "image" / name
        label_vessel_nii = output_ssm_nii__with_vessels_root / case_dir.name / "label" / name
        
        shape_morph_case_dir = shape_morph_root / case_dir.name
        shape_morph_image_nii = shape_morph_case_dir / "image" / name
        shape_morph_label_nii = shape_morph_case_dir / "label" / name
        shape_morph_coronary_nii = shape_morph_case_dir / "coronary" / name

        shape_morph_image_nii.parent.mkdir(exist_ok=True, parents=True)
        shape_morph_image_nii.symlink_to(image_vessel_nii)
        shape_morph_label_nii.parent.mkdir(exist_ok=True, parents=True)
        shape_morph_label_nii.symlink_to(label_nii)

        label_vessel_image = nib.load(label_vessel_nii)
        affine = label_vessel_image.affine
        label_vessel_data = label_vessel_image.get_fdata()
        coronary_art_mask = label_vessel_data == 6
        coronary_vein_mask = label_vessel_data == 7
        new_array = np.zeros_like(label_vessel_data)
        new_array[coronary_art_mask] = 1
        new_array[coronary_vein_mask] = 2

        coronary_art_nii = nib.Nifti1Image(new_array, affine)
        shape_morph_coronary_nii.parent.mkdir(exist_ok=True, parents=True)
        coronary_art_nii.to_filename(shape_morph_coronary_nii)
        
        
```
